```r
#1. Calculate the following:
##a. (5^4) + (5*(5^3)) + ... + (100^4) + 5*(100^3))
#With Loop
print("With Loop")
a <- 5
Pre <- 0
while (a <= 100){
    Cur <- ((a^4) + (5*(a^3))) + Pre
    Pre <- Cur
    a <- a+1
}
Cur
#Without Loop
print("Without Loop")
a <- 5:100
A <- sum((a^4) + 5 * (a^3))
A
```

```r
##b. Calculate U40
n <- 3
Pre <- 2
PrePre <- 1
while (n <= 40){
    Cur <- Pre + PrePre
    PrePre <- Pre
    Pre <- Cur
    n <- n + 1
}
U40 <- Cur
U40
```

```r
##c. Calculate Cos x and Sin x (Taylor Series)
n <- 0:45
cat("<<Cos(x) and Sin(x)>>\nEnter x value: ")
x <- as.integer(readline())
y <- (x*pi)/180
cos <- sum(((-1)^n)*((y)^(2*n)/factorial((2*n))))
sin <- sum(((-1)^n)*((y)^((2*n)+1)/factorial((2*n)+1)))
cat("Cos(",x,"): ",cos,"\n")
cat("Sin(",x,"): ",sin)
```

```r
##d. Create Vetor of value e^x Cos(x):
x <- 3.0
v <- c()
while (x <= 6){
    A <- c(exp(x) * cos(x))
    v <- c(v, A)     
    x <- x + 0.1      
}
v
```

```r
##e. Divisable by 2:
x <- 1:300
check <- 1
count <- 0
while (check <= length(x)){
    if(x[check] %% 2 == 0){
        count <- count + 1
    }
    check <- check + 1
}
count
```

```r
##c. Find missing value and replace them with mean(Tidy data)
library(dataMaid)
covid_19_data <- read.csv(file = "../input/covid19-in-turkey/covid_19_data_tr.csv")
makeDataReport(covid_19_data, replace=TRUE)
```

```r
##d. Generate Means and Median
i <- 1
c <- c()
d <- c()
r <- c()
while (i <= nrow(covid_19_data)){
    c <- c(v, covid_19_data[i,"Confirmed"])
    d <- c(v, covid_19_data[i,"Deaths"])
    r <- c(v, covid_19_data[i,"Recovered"])
    i <- i + 1
}
cat("Mean of Confirmed cases: ",mean(c),"\n")
cat("Mean of Deaths : ",mean(d),"\n")
cat("Mean of Recovery : ",mean(r),"\n\n")
cat("Median of Confirmed cases: ",median(c),"\n")
cat("Median of Deaths : ",median(d),"\n")
cat("Median of Recovery : ",median(r),"\n")
```

```r
##e. Rename 2 existing variable:

colnames(covid_19_data)[3] <- "Confirmed_Cases"
colnames(covid_19_data)[6] <- "Date"
colnames(covid_19_data)
```

```r
##f. Create Plot using 7 layers grammar of graphic:
library(ggplot2)
case <- covid_19_data[["Confirmed_Cases"]]
death <- covid_19_data[["Deaths"]]
recover <- covid_19_data[["Recovered"]]
day <- 1:nrow(covid_19_data)
##i. Set axis
##ii. Specify parameter
head(covid_19_data)
plot_case <- ggplot(covid_19_data, mapping= aes(day, y = case)) + geom_point(colour = "red", alpha = 0.2, size = 5)
plot_case
plot_death <- ggplot(covid_19_data, mapping= aes(day, y = death)) + geom_point(colour = "black", alpha = 0.2, size = 5)
plot_death
plot_recover <- ggplot(covid_19_data, mapping= aes(day, y = recover)) + geom_point(colour = "green", alpha = 0.2, size = 5)
plot_recover
```
